# 2. Data description

File `/ datasets / hypothesis.csv` (9 hypotheses for increasing online store revenue):
 - `Hypothesis` - a short description of the hypothesis;
 - `Reach` - user coverage on a 10-point scale;
 - `Impact` - impact on users on a 10-point scale;

File `/ datasets / visitors.csv`:
- `date` - date;
- `group` - group of A / B-test;
- `visitors` - the number of users on the specified date in the specified A / B-test group

# 5. Hypothesis prioritizing 
<a id='Приоритезация гипотез'></a>

**ICE Framework**



$$ICE = \frac{Impact * Confidence}{Efforts}$$



```python
hypothesis['ICE'] = hypothesis['impact'] * hypothesis['confidence'] / hypothesis['efforts']
```

```python
hypothesis.sort_values(by = 'ICE', ascending = False)
```

**RICE Framework**


$$RICE = \frac{Rearch * Impact * Confidence}{Efforts}$$

```python
hypothesis['RICE'] = hypothesis['reach'] * hypothesis['impact'] * hypothesis['confidence'] / hypothesis['efforts']
```

```python
hypothesis.sort_values(by = 'RICE', ascending = False)
```

```python
for i in enumerate(hypothesis['hypothesis']):
    print(i[0], i[1])
```

    - # 8 Launch a promotion giving a discount on goods on your birthday
    - # 0 Add two new channels to attract traffic, which will attract 30% more users
    - # 7 Add a subscription form to all main pages to build a customer base for email newsletters

The RICE method distinguishes the following hypotheses in the top-3:

    - # 7 Add a subscription form to all main pages to build a customer base for email newsletters
    - # 2 Add blocks of product recommendations to the website of the online store to increase conversion and average order value
    - # 0 Add two new channels to attract traffic, which will attract 30% more users

Thus, hypotheses # 7 and # 0 (the most significant) in terms of product improvement. The framework difference is due to the reach parameter - which we characterize the audience reach. **Build a graph of cumulative revenue by group. Draw conclusions and assumptions.**

```python
# To plot graphs, you need to collect cumulative data. Let's declare the dataframe cumulativeData
# create an array of unique pairs of date values and test groups
datesGroups = orders[['date','group']].drop_duplicates()

# Get the rows of the orders table whose date is less than or equal to the date of the item from datesGroups, and the test group is equal to the group from datesGroups:
ordersAggregated = datesGroups.apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])]\
.agg({'date' : 'max', 
      'group' : 'max', 
      'transactionId' : pd.Series.nunique, 
      'visitorId' : pd.Series.nunique, 
      'revenue' : 'sum'}), axis=1).sort_values(by=['date','group']) 

# Similarly, we get the aggregated cumulative daily data on visitors:
visitorsAggregated = datesGroups.apply(lambda x: visitors[np.logical_and(visitors['date'] <= x['date'], visitors['group'] == x['group'])]\
                                       .agg({'date' : 'max', 
                                             'group' : 'max', 
                                             'visitors' : 'sum'}), axis=1).sort_values(by=['date','group'])

# combine cumulative data in one table and assign clear names to its columns
cumulativeData = ordersAggregated.merge(visitorsAggregated, left_on=['date', 'group'], right_on=['date', 'group'])
cumulativeData.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']

display(cumulativeData.head(5)) 
```

```python
# Let's build graphs of cumulative revenue by days and groups of A / B testing
# dataframe with cumulative number of orders and cumulative revenue by day in group A
cumulativeRevenueA = cumulativeData[cumulativeData['group']=='A'][['date','revenue', 'orders']]

# dataframe with cumulative number of orders and cumulative revenue by day in group B
cumulativeRevenueB = cumulativeData[cumulativeData['group']=='B'][['date','revenue', 'orders']]

# Build a revenue graph
fig = plt.figure();
ax = plt.subplots(figsize=(18, 8));

# Plotting the revenue of group A
plt.plot(cumulativeRevenueA['date'], 
         cumulativeRevenueA['revenue'], 
         label='A', 
         linewidth = 3);

# Plotting the revenue of group B
plt.plot(cumulativeRevenueB['date'], 
         cumulativeRevenueB['revenue'], 
         label='B',
         linewidth = 3);
plt.legend();

# Name the graph
plt.title("Revenue CUMSUM Graph", fontsize = 20);
plt.xlabel("Day", fontsize = 20); # label for x-axis
plt.ylabel("Revenue", fontsize = 20); # label for y-axis
```

**Conclusions:**
    
There is a strong jump in revenue from 17 to 21 August for group B. This may signal spikes in the number of orders, or the appearance of very expensive orders in the sample. Group B leads in terms of revenue over the entire period of time.

**Calculate the statistical significance of differences in conversion between groups based on raw data. Draw conclusions and make assumptions.**


```python
visitorsADaily = visitors[visitors['group']=='A'][['date', 'visitors']]
visitorsADaily.columns = ['date', 'visitorsPerDateA']

visitorsACummulative = visitorsADaily.apply(lambda x: visitorsADaily[visitorsADaily['date'] <= x['date']] \
                                            .agg({'date' : 'max', 'visitorsPerDateA' : 'sum'}), axis=1)
visitorsACummulative.columns = ['date', 'visitorsCummulativeA']

visitorsBDaily = visitors[visitors['group']=='B'][['date', 'visitors']]
visitorsBDaily.columns = ['date', 'visitorsPerDateB']

visitorsBCummulative = visitorsBDaily.apply(lambda x: visitorsBDaily[visitorsBDaily['date'] <= x['date']] \
                                            .agg({'date' : 'max', 'visitorsPerDateB' : 'sum'}), axis=1)
visitorsBCummulative.columns = ['date', 'visitorsCummulativeB']
```

```python
ordersADaily = orders[orders['group']=='A'][['date', 'transactionId', 'visitorId', 'revenue']]\
    .groupby('date', as_index=False)\
    .agg({'transactionId' : pd.Series.nunique, 'revenue' : 'sum'})
ordersADaily.columns = ['date', 'ordersPerDateA', 'revenuePerDateA']

ordersACummulative = ordersADaily.apply(
    lambda x: \
        ordersADaily[ordersADaily['date'] <= x['date']]\
            .agg({
                'date' : 'max',
                'ordersPerDateA' : 'sum',
                'revenuePerDateA' : 'sum'}), axis=1).sort_values(by=['date'])
ordersACummulative.columns = ['date', 'ordersCummulativeA', 'revenueCummulativeA']

ordersBDaily = orders[orders['group']=='B'][['date', 'transactionId', 'visitorId', 'revenue']]\
    .groupby('date', as_index=False)\
    .agg({'transactionId' : pd.Series.nunique, 'revenue' : 'sum'})
ordersBDaily.columns = ['date', 'ordersPerDateB', 'revenuePerDateB']

ordersBCummulative = ordersBDaily.apply(
    lambda x: \
        ordersBDaily[ordersBDaily['date'] <= x['date']]\
            .agg({
                'date' : 'max',
                'ordersPerDateB' : 'sum',
                'revenuePerDateB' : 'sum'}), axis=1).sort_values(by=['date'])
ordersBCummulative.columns = ['date', 'ordersCummulativeB', 'revenueCummulativeB']
```

```python
data = ordersADaily.merge(ordersBDaily, left_on='date', right_on='date', how='left')\
    .merge(ordersACummulative, left_on='date', right_on='date', how='left')\
    .merge(ordersBCummulative, left_on='date', right_on='date', how='left')\
    .merge(visitorsADaily, left_on='date', right_on='date', how='left')\
    .merge(visitorsBDaily, left_on='date', right_on='date', how='left')\
    .merge(visitorsACummulative, left_on='date', right_on='date', how='left')\
    .merge(visitorsBCummulative, left_on='date', right_on='date', how='left')

display(data.head(5)) 
```

data column names:

- `ordersCummulativeB` - total number of orders up to and including the selected date in group B;

- `visitorsCummulativeA` - the number of users before the selected date inclusive in group A;
- `visitorsCummulativeB` - the number of users before the selected date inclusive in group B.

```python
# Let's calculate the statistical significance of the difference in conversion between groups
ordersByUsersA = orders[orders['group']=='A'].groupby('visitorId', as_index=False).agg({'transactionId' : pd.Series.nunique})
ordersByUsersA.columns = ['userId', 'orders']

ordersByUsersB = orders[orders['group']=='B'].groupby('visitorId', as_index=False).agg({'transactionId' : pd.Series.nunique})
ordersByUsersB.columns = ['userId', 'orders']
```

```python
# Let's declare the variables sampleA and sampleB, in which the number of orders will correspond to users from different groups.
# Those who have not ordered anything will be matched with zeros. This is necessary to prepare samples for testing by the Mann-Whitney criterion.

sampleA = pd.concat([ordersByUsersA['orders'],
                     pd.Series(0, index=np.arange(data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])), 
                               name='orders')],axis=0)

sampleB = pd.concat([ordersByUsersB['orders'],
                     pd.Series(0, index=np.arange(data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])), 
                               name='orders')],axis=0) 
```

```python
ordersByUsersA = orders[orders['group']=='A'].groupby('visitorId', as_index=False).agg({'transactionId' : pd.Series.nunique})
ordersByUsersA.columns = ['userId', 'orders']

ordersByUsersB = orders[orders['group']=='B'].groupby('visitorId', as_index=False).agg({'transactionId' : pd.Series.nunique})
ordersByUsersB.columns = ['userId', 'orders']

sampleA = pd.concat([ordersByUsersA['orders'],
                     pd.Series(0, index=np.arange(data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])), name='orders')],axis=0)

sampleB = pd.concat([ordersByUsersB['orders'],
                     pd.Series(0, index=np.arange(data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])), name='orders')],axis=0)
```

**Let's check the hypotheses:**

  - `Ho`: there are no statistically significant differences in conversion rates between group A and group B

  - `H1`: there are statistically significant differences in conversion rates between group A and group B


```python
# Assume that there are no statistically significant differences in conversion rates between group A and group B. This will be our hypothesis Ho.
# Then the alternative hypothesis H1 - the difference in conversion between groups is
# set the level of statistical significance alpha = 0.05
# if the P-value is greater than alpha, then there is no reason to reject the null hypothesis

alpha = 0.05

if (stats.mannwhitneyu(sampleA, sampleB)[1] < alpha):
    print("Reject Ho hypothesis => There is a difference in conversions between groups")
else:
    print("Can not reject Ho hypothesis => There is no difference in conversion")
    
print("P-value = {0:.3f}".format(stats.mannwhitneyu(sampleA, sampleB)[1]))
print("The relative average conversion of group B is higher by {0:.3f}".format(sampleB.mean()/sampleA.mean()-1))
```

**Conclusion:**

Hypothesis `H0` is rejected. According to the raw data, there is a difference in conversion between groups A and B. If we “clean up” the data, it is possible that the results will change.

**Calculate the statistical significance of the differences in the average order receipt between groups according to the "raw" data. Draw conclusions and make assumptions.**

```python
# find the relative differences in the average check between groups
# Ho - there are differences in the average check between group A and group B
# H1 - there is no difference between group A and group B on the average bill

alpha = 0.05

if (stats.mannwhitneyu(orders[orders['group']=='A']['revenue'], orders[orders['group']=='B']['revenue'])[1] < alpha):
    print("Reject Ho hypothesis => there is no difference in the average bill between group A and group B")
else:
    print("Can not reject Ho hypothesis => there are differences in the average check between group A and group B")

print("P-value = {0:.3f}".format(stats.mannwhitneyu(orders[orders['group']=='A']['revenue'], 
                                                    orders[orders['group']=='B']['revenue'])[1]))
print("Relative differences in average check between groups = {0:.3f}".format(orders[orders['group']=='B']['revenue'].mean()/orders[orders['group']=='A']['revenue'].mean()-1)) 
```

**Conclusion:**

According to the raw data, there is a difference in conversion between groups A and B. If we “clean up” the data, it is possible that the results will change.

We cannot say that the difference is statistically significant. No statistically significant differences were found, since P-value is high.
The difference in checks can be an accident.
