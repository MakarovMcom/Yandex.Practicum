## Data Processing
<a id='Обработка данных'></a>

```python
#create new DF to avaid changes in intial data
games_data = df

# change the names in new DF
games_data.columns = games_data.columns.str.lower()

# we have 2 strings where value = NaN in the column 'name'. So we can delete it without hesitation
games_data = games_data.dropna(subset=['name'])

# lets have a look at strings with empty year
display(games_data.query('year_of_release != year_of_release').head())
len_frame = len(games_data)
len_empty_year = len(games_data.query('year_of_release != year_of_release'))
print("we've got empty YEAR strings in percent - {:.1%}".format(len_empty_year/len_frame))
```

```python
# the problem is there is no logic for filling the year with a strong logic that is why we have two 
# options:
# 1. delete and that quite easy because the total number is tiny
# 2. fill with a values which is prefereble 

print()
print('year_of_release before fill =', games_data['year_of_release'].isna().sum())
games_data = games_data.sort_values(['platform', 'genre'])
games_data['year_of_release'] = games_data['year_of_release'].fillna(method = 'ffill')
print('year_of_release after fill =', games_data['year_of_release'].isna().sum())

# change the dtype year_of_release to int 
games_data['year_of_release'] = games_data['year_of_release'].astype(int)
display(games_data.sample(5))
```

### Raitings and scores
<a id='Рейтинги и оценки'></a>

We have empty spaces in colums `critic_score` and `user_score`. At the first look from the sample above in has correlation. Let's find out how strong is it and do we have a probability to fill one column from another. That is why we have to make a small survey in a group which doesn't have emprty spaces.

```python
# lets have a look at critic_score
print('critic_score before fillings =', games_data['critic_score'].isna().sum())
print(games_data['critic_score'].unique())
```

```python
# lets have a look at user_score
print('user_score before fillings =', games_data['user_score'].isna().sum())
print(games_data['user_score'].unique())
```

```python
print()
critic_score_not_na = games_data.loc[~(games_data['critic_score'].isna())]
critic_score_not_na = critic_score_not_na.loc[:,'name':'critic_score']

user_score_not_na = games_data.loc[(~(games_data['user_score'].isna())) & \
                                   (games_data['user_score'] != 'tbd')]
user_score_not_na = user_score_not_na.loc[:,'user_score']

scores = critic_score_not_na.join(user_score_not_na)
scores = scores.dropna(subset=['user_score'])

scores['user_score'] = scores['user_score'].astype(float)

print('Our correlation of Pirsone is: {:.3f}'
                                              .format(scores['critic_score']
                                              .corr(scores['user_score']))
     )
```

```python
scores.plot(
    x = 'user_score', 
    y = 'critic_score', 
    kind = 'scatter',
    grid = 'True',
    alpha = 0.6,
    title = 'Correlation graph between user_score & critic_score',
    figsize = (15,6));
```

To make a conclusion we have strong correlation between user_score and critic_score in values from 5 to 9.
We needed to translate from user score to critic score and back. That is the reason to create ration.

```python
# from user_score to critic_score ratio
above_5 = scores.loc[scores['user_score'] >= 5]
ratio_5 = above_5['critic_score'] / above_5['user_score']
ratio_5 = ratio_5.median()
print('for (user_score >=5) ratio is {:.2f}'.format(ratio_5))

above_50 = scores.loc[scores['critic_score'] >= 50]
ratio_50 = above_50['critic_score'] / above_50['user_score']
ratio_50 = ratio_50.median()
print('for (critic_score >=50) ratio is {:.2f}'.format(ratio_50))
```

```python
score_series = scores.loc[:,['critic_score', 'user_score']]
games_data = games_data.join(score_series, rsuffix = '_y')
display(games_data)
```

```python
display(games_data.loc[(games_data['user_score_y'] >= 5) & \
        (games_data['critic_score'].isna()), 'critic_score'])

display(games_data.loc[(~games_data['user_score_y'].isna()) & \
        (games_data['critic_score'].isna()), 'critic_score'])

```

But the dataframe is empty and it's not a good option, so come back to games_data, delete new columns with **suffixs _y** and fill empty cells by extra score None to point them without changing statistics parametrs:

At the first **`Yandex`** search link on the `request = "e10+ rating"` link we can find that ratings is content restriction in the games. 

https://ru.wikipedia.org/wiki/Entertainment_Software_Rating_Board

To have better understanding place the standarts below:

`'E'` - everyone, all ages [0 and above]

`'E10+'` - everyone 10 + [10 and above]

`'M'` - mature [17 and above]

`'T'` - teen [13 and above]

`'EC'` - early childhood [from 0 to 7]

`'RP'` - rating pending, without final rating [None]

`'K-A'` - Kids to Adults, same as everyone [0 and above] 

`'AO'` - Adults Only [18 and above]

Clarify the standarts in DataFrame:
1. replace `'K-A'` to `'E'`


```python
games_data.loc[games_data['rating'] == 'K-A', 'rating'] = 'E'
```

```python
df = games_data.loc[games_data['rating'].isna()]
pivot_df = df.pivot_table(index ='genre', values = 'name', aggfunc = 'count')

display(pivot_df)
```

Make the manual classification:

Action = `'M'`

Adventure = `'T'`

Fighting = `'M'`

Misc = `'RP'`

Platform = `'RP'`

Puzzle = `'E'`

Racing = `'E'`

Role-Playing = `'RP'`

Shooter = `'M'`

Simulation = `'RP'`

Sports = `'E10+'`

Strategy = `'T'`

```python
def replace(row):
    if row['genre'] == 'Action':
        return 'M'
    elif row['genre'] == 'Adventure':
        return 'T'
    elif row['genre'] == 'Fighting':
        return 'M'
    elif row['genre'] == 'Misc':
        return 'RP'
    elif row['genre'] == 'Platform':
        return 'RP'
    elif row['genre'] == 'Puzzle':
        return 'E'
    elif row['genre'] == 'Racing':
        return 'E'
    elif row['genre'] == 'Role-Playing':
        return 'RP'
    elif row['genre'] == 'Shooter':
        return 'M'
    elif row['genre'] == 'Simulation':
        return 'RP'
    elif row['genre'] == 'Sports':
        return 'E10+'
    elif row['genre'] == 'Strategy':
        return 'T'
    else:
        return "RP"
    
games_data['rating'] = games_data.apply(replace, axis = 1)    
```

**How does user and critic reviews affect sales within one popular platform?**

```python
# correlation between all platforms and total_sale
df_scores = games_data.loc[:,['critic_score', 'user_score', 'total_sales']]
display(df_scores.corr())
```

```python
# correlation between PS4 platform and total_sale
df_scores_PS4 = games_data.query('platform == "PS4"').loc[:,['critic_score', 'user_score', 'total_sales']]
display(df_scores_PS4.corr())
```

Evaluating the `critic_score` and `user_score` for all platforms according to Pearson's coefficient, we can conclude that there is a relationship between them at high ratings `(user_score> = 5)`. The influence of `critic_score` and `user_score` on total sales is insignificant, most likely due to the large number of outliers in the upper border. Let's check the assumption.

```python
# make a threshold to cut emissions
df_scores_threshold = games_data['total_sales'].quantile(0.9)
df_scores = games_data.query('total_sales <= @df_scores_threshold').loc[:,['platform','critic_score', 'user_score', 'total_sales']]
for platform in df_scores['platform'].unique():
    corr_value = df_scores.query('platform == @platform')['total_sales'].corr(df_scores['critic_score'])
    print("Платформа {} имеет корреляцию {:.2f} с продажами".format(platform, corr_value))

display(df_scores.corr())
```

Despite the fact that we got rid of some of the outliers, the situation with the dependence of `critic_score` on `total_sales` practically does not change, i.e. connection is insignificant.

## Hypothesis testing
<a id='Проверка гипотез'></a>

- Average user ratings for Xbox One and PC platforms are the same;
- The average user ratings for the genres Action (action games) and Sports are different.

Zero hypothesis. `Ho`: Average user ratings for XOne and PC platforms are the same

Alternative hypothesis. `H1`: Average user ratings for XOne and PC platforms differ

```python
from scipy import stats as st
import numpy as np
       
sample_1 = games_data.query('platform == "XOne"').loc[:,'user_score'] 
sample_2 = games_data.query('platform == "PC"').loc[:,'user_score']
```

```python
fig = plt.figure()

ax_1 = fig.add_subplot(2, 2, 1)
ax_2 = fig.add_subplot(2, 2, 2)

ax_1.hist(sample_1, bins='auto')
ax_2.hist(sample_2, bins='auto')

ax_1.set_title('Allocation User_Score for XOne')
ax_2.set_title('Allocation User_Score for PC')

ax_1.set_xlabel('User_score')
ax_2.set_xlabel('User_score')

ax_1.set_ylabel('Frequency')
ax_1.set_ylabel('Frequency')

plt.show()
```

```python
# check the variance of sample_1 and sample_2 to set equal_var = True
variance_1 = np.var(sample_1)
variance_2 = np.var(sample_2)
print('var_1: {:.3f}'.format(variance_1))
print('var_2: {:.3f}'.format(variance_2))
print('Difference of var1 and var_2 = {:.0%}'.format(variance_1 / variance_2))
print()

alpha = 0.05

results = st.ttest_ind(sample_1.dropna(), sample_2.dropna(), equal_var = False)
print('p-value: {:.6f}'.format(results.pvalue))

if results.pvalue >= alpha:
    print("Can not reject Ho hypothesis")
else:
    print("Reject Ho hypothesis")
```

```python
# verification
print('Средние пользовательские рейтинги XOne - {}'.format(sample_1.dropna().median()))
print('Средние пользовательские рейтинги PC - {}'.format(sample_2.dropna().median()))
```

Zero hypothesis. `Ho`: Average user ratings for Action and Sports genres are different

Alternative hypothesis. `H1`: Average user ratings for Action and Sports genres are the same

```python
sample_3 = games_data.query('genre == "Action"').loc[:,'user_score'] 
sample_4 = games_data.query('genre == "Sports"').loc[:,'user_score']
```

```python
fig = plt.figure()

ax_1 = fig.add_subplot(2, 2, 1)
ax_2 = fig.add_subplot(2, 2, 2)

ax_1.hist(sample_3, bins='auto')
ax_2.hist(sample_4, bins='auto')

ax_1.set_title('Allocation User_Score for Action')
ax_2.set_title('Allocation User_Score for Sports')

ax_1.set_xlabel('User_score')
ax_2.set_xlabel('User_score')

ax_1.set_ylabel('Frequency')
ax_1.set_ylabel('Frequency')

plt.show()
```

```python

# check the variance of sample_1 and sample_2 to set equal_var = True
variance_3 = np.var(sample_3)
variance_4 = np.var(sample_4)
print('var_3: {:.3f}'.format(variance_3))
print('var_4: {:.3f}'.format(variance_4))
print('Difference of var1 and var_2 = {:.0%}'.format(variance_1 / variance_2))
print()

alpha = 0.05

results_2 = st.ttest_ind(sample_1.dropna(), sample_2.dropna(), equal_var = False)
print('p-value: {:.6f}'.format(results_2.pvalue))

if (results_2.pvalue < alpha):
    print("Reject Ho hypothesis")
else:
    print("Can not reject Ho hypothesis")
```

```python
# verification
print('Средние пользовательские рейтинги Action - {}'.format(sample_3.dropna().median()))
print('Средние пользовательские рейтинги Sports - {}'.format(sample_4.dropna().median()))
```

## Conclusion: Hypothesis testing
<a id='Выводы: Проверка гипотез'></a>

When testing hypotheses, the rule about equality of the means of two general populations was used. The parameter is equal_var = False, since the variances of the samples are very different. The level of statistical significance is set at 5%.

The hypothesis that the average user ratings of the XOne and PC platforms are the same was rejected.
Therefore, the average user ratings differ, which is confirmed by calculations:
- average user rating XOne - 6.8
- average user rating PC - 7.5

The hypothesis that the average user ratings for the Action and Sports genres are different was rejected.
Therefore, the average user is equal, which is confirmed by calculations:
- average user Action ratings - 7.4
- Average Sports User Ratings - 7.4
