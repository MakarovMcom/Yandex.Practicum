## Описание данных
<a id='Описание данных'></a>

**Таблица users (информация о пользователях):**

/datasets/users.csv — расположение файла
- user_id — уникальный идентификатор пользователя
- first_name — имя пользователя
- last_name — фамилия пользователя
- age — возраст пользователя (годы)
- reg_date — дата подключения тарифа (день, месяц, год)
- churn_date — дата прекращения пользования тарифом (если значение пропущено, то тариф ещё действовал на момент выгрузки данных)
- city — город проживания пользователя
- tariff — название тарифного плана

**Таблица calls (информация о звонках):**

/datasets/calls.csv  — расположение файла
- id — уникальный номер звонка
- call_date — дата звонка
- duration — длительность звонка в минутах
- user_id — идентификатор пользователя, сделавшего звонок

**Таблица messages (информация о сообщениях):**

/datasets/messages.csv — расположение файла
- id — уникальный номер сообщения
- message_date — дата сообщения
- user_id — идентификатор пользователя, отправившего сообщение

**Таблица internet (информация об интернет-сессиях):**

/datasets/internet.csv — расположение файла
- id — уникальный номер сессии
- mb_used — объём потраченного за сессию интернет-трафика (в мегабайтах)
- session_date — дата интернет-сессии
- user_id — идентификатор пользователя

**Таблица tariffs (информация о тарифах):**

/datasets/tariffs.csv — расположение файла
- tariff_name — название тарифа
- rub_monthly_fee — ежемесячная абонентская плата в рублях
- minutes_included — количество минут разговора в месяц, включённых в абонентскую плату
- messages_included — количество сообщений в месяц, включённых в абонентскую плату
- mb_per_month_included — объём интернет-трафика, включённого в абонентскую плату (в мегабайтах)
- rub_per_minute — стоимость минуты разговора сверх тарифного пакета (например, если в тарифе 100 минут разговора в месяц, то со 101 минуты будет взиматься плата)
- rub_per_message — стоимость отправки сообщения сверх тарифного пакета
- rub_per_gb — стоимость дополнительного гигабайта интернет-трафика сверх тарифного пакета (1 гигабайт = 1024 мегабайта)


**Важно!** «Мегалайн» всегда округляет вверх значения минут и мегабайтов. Если пользователь проговорил всего 1 секунду, в тарифе засчитывается целая минута.

## Обработка данных

### Исправление типов данных, заполнение пропусков

```python
# skips before filling
print('Skips before filling churn_date:',  user_data['churn_date'].isna().sum())

# churn_data to date format
user_data['churn_date'] = pd.to_datetime(user_data['churn_date'], format = '%Y-%m-%d')
user_data['reg_date'] = pd.to_datetime(user_data['reg_date'], format = '%Y-%m-%d')

#if there are skips then assign the value 2018-12-31 
user_data.loc[user_data['churn_date'].isna(), 'churn_date'] = pd.Timestamp(2018, 12, 31)

print('Skips after filling churn_date:',  user_data['churn_date'].isna().sum())

# check out the result
display(user_data.sample(5))
```

```python
print(user_data.info())
```

```python
# call_date to date format
calls_data['call_date'] = pd.to_datetime(calls_data['call_date'], format = '%Y-%m-%d')

import numpy as np 

def round_cell(col):
    if col == 0:
        return col
    else:
        col = np.ceil(col)
        return col

calls_data['duration'] = calls_data['duration'].apply(round_cell)   
display(calls_data.sample(5))
```

```python
# message_date to date format
mes_data['message_date'] = pd.to_datetime(mes_data['message_date'], format = '%Y-%m-%d')
display(mes_data.sample(5))
```

```python
# session_date to date format
int_data['session_date'] = pd.to_datetime(int_data['session_date'], format = '%Y-%m-%d')

# round the internet traffic
int_data['mb_used'] = int_data['mb_used'].apply(round_cell)  

# Unnamed: 0 column doesnt exist in the discription, so we can delete this column
int_data = int_data.drop(columns = ['Unnamed: 0'])
display(int_data.sample(5))
```

```python
# let's have a look how many call we have where duration = 0
dur_null = calls_data.query('duration == 0')
dur_not_null = calls_data.query('duration != 0')
percent_null = len(dur_null) / len(calls_data)

print("Number of call with duration=0 is : {}".format(len(dur_null)))
print("Total number of calls is : {}".format(len(calls_data)))
print("Number of call with duration=0 is : {:.1%}".format(percent_null))
```

```python
num_users_with_nulls = len(dur_null.pivot_table(index = 'user_id', values = 'duration', aggfunc = 'count'))
print('Number of users which have calls with duration = 0 is :',num_users_with_nulls)
```

There are too many users (492 from 500) who have `duration = 0`, can not drop this data out

### расчет значений

```python
tariff_data['gb_per_month_included'] = tariff_data['mb_per_month_included'] / 1024
```

```python
# number of calls and minutes of conversation by months and user_id
calls_data['month'] = calls_data['call_date'].dt.month
pivot_table_calls = calls_data.pivot_table(index = ['user_id', 'month'], values = "duration", aggfunc = ['count', 'sum'])

# rename our columns
pivot_table_calls = pivot_table_calls.reset_index()
pivot_table_calls.columns = ['user_id', 'month', 'num_of_calls', 'sum_of_calls']
display(pivot_table_calls.head(5))
display(pivot_table_calls.columns)
```

```python
# number of messages by months and user_id
mes_data['month'] = mes_data['message_date'].dt.month
mes_data['num'] = 1
pivot_table_mes = mes_data.pivot_table(index = ['user_id', 'month'], values = "num", aggfunc = ['sum'])
pivot_table_mes = pivot_table_mes.reset_index()
pivot_table_mes.columns = ['user_id', 'month', 'num_of_mes']
display(pivot_table_mes.head(5))
```

```python
# amount of traffic by months and user_id
int_data['month'] = int_data['session_date'].dt.month
int_data['gb'] = int_data['mb_used'] / 1024
#int_data['gb'] = int_data['gb'].apply(round_cell)
pivot_table_int = int_data.pivot_table(index = ['user_id', 'month'], values = "gb", aggfunc = ['sum'])
pivot_table_int = pivot_table_int.reset_index()
pivot_table_int.columns = ['user_id', 'month', 'sum_of_gb']
pivot_table_int['sum_of_gb'] = pivot_table_int['sum_of_gb'].apply(round_cell)
display(pivot_table_int.head(5))
```

<a id='гиги'></a>

```python
# merge tables
full_data_1 = user_data.merge(pivot_table_calls, 
                              on = 'user_id', how = 'left')
full_data_2 = full_data_1.merge(pivot_table_mes, 
                                on = ['user_id', 'month'], how = 'left')
full_data = full_data_2.merge(pivot_table_int, 
                              on = ['user_id', 'month'], how = 'left')
display(full_data.head(10))
#display(full_data_1.head(10))
#display(full_data_2.head(10))
```

```python
full_data.info()
```

```python
# let's have a look at data which has NaN in column Month
# if month = NaN it means that we didn't have calls messages and internet sessions in that period
# because the customer decided to quit from this service
# fill the gaps with null
display(full_data.query('month != month'))
print("Total number of strings with Month = Nan :",len(full_data.query('month != month')))
```

```python
full_data.loc[full_data['month'].isna(), ['num_of_calls', 'sum_of_calls', 'num_of_mes', 
                                          'sum_of_gb', 'month']] = 0
print("Total number of strings with Month = Nan :",len(full_data.query('month != month')))
```

```python
# the same thing in columns num_of_mes and sum_of_gb 
# if we have Nan seems that user didn't used this services
# fill the gaps with null

full_data[['num_of_mes', 'sum_of_gb']] = full_data[['num_of_mes', 'sum_of_gb']].fillna(value=0)
```

```python
print(full_data.info())
display(full_data)
```

```python
# add information about tariffs in the table
full_data.rename(columns={'tariff': 'tariff_name'}, inplace = True)
full_data = full_data.merge(tariff_data, on = 'tariff_name', how = 'left')
display(full_data.head())
```

```python
full_data['rest_min'] = full_data['minutes_included'] - full_data['sum_of_calls']
full_data['rest_mes'] = full_data['messages_included'] - full_data['num_of_mes']
full_data['rest_gb'] = full_data['gb_per_month_included'] - full_data['sum_of_gb']

# if rest_min >= 0 then we don't have extra payment and rest_min = 0 
# if rest_mes >= 0 then we don't have extra payment and rest_mes = 0
# if rest_gb >= 0 then we don't have extra payment and rest_mes = 0 

full_data.loc[full_data['rest_min'] >= 0, 'rest_min'] = 0
full_data.loc[full_data['rest_mes'] >= 0, 'rest_mes'] = 0
full_data.loc[full_data['rest_gb'] >= 0, 'rest_gb'] = 0

# if rest_min < 0 then rest_min = abs(rest_min) 
# if rest_mes < 0 then rest_mes = abs(rest_min)
# if rest_gb < 0 then rest_mes = abs(rest_min)

full_data.loc[full_data['rest_min'] < 0, 'rest_min'] = abs(full_data['rest_min'])
full_data.loc[full_data['rest_mes'] < 0, 'rest_mes'] = abs(full_data['rest_mes'])
full_data.loc[full_data['rest_gb'] < 0, 'rest_gb'] = abs(full_data['rest_gb'])

full_data['revenue'] = full_data['rub_monthly_fee'] + \
                    full_data['rest_min'] * full_data['rub_per_minute'] + \
                    full_data['rest_mes'] * full_data['rub_per_message'] + \
                    full_data['rest_gb'] * full_data['rub_per_gb']
display(full_data.head())
```

## Проверка гипотез
<a id='Проверка гипотез'></a>

- средняя выручка пользователей тарифов «Ультра» и «Смарт» различается;
- средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов;

```python
from scipy import stats as st
import numpy as np
       
sample_1 = full_data.query('tariff_name == "smart"').loc[:,'revenue'] 
sample_2 = full_data.query('tariff_name == "ultra"').loc[:,'revenue']

# check the variance of sample_1 and sample_2 to set equal_var = True
variance_1 = np.var(sample_1, ddof = 1)
variance_2 = np.var(sample_2, ddof = 1)
print('var_1: {:.3f}'.format(variance_1))
print('var_2: {:.3f}'.format(variance_2))
print('Difference of var1 and var_2 = {:.0%}'.format(variance_1 / variance_2))
print()

#std_1 = np.std(sample_1, ddof = 1)
#std_2 = np.std(sample_2, ddof = 1)
#print('std_1: {:.3f}'.format(std_1))
#print('std_2: {:.3f}'.format(std_2))


alpha = 0.05

results = st.ttest_ind(sample_1, sample_2, equal_var = False)
print('p-value: {:.3f}'.format(results.pvalue))

# Нулевая гепотеза. Ho: Средняя выручка по обоим тарифам равна
# Альтернативная гипотеза. H1: Среняя выручка по тарифам отличается
# Zero hypothesis. Ho: The average revenue for both tariffs is
# Alternative hypothesis. H1: Average tariff revenue is different

if (results.pvalue < alpha):
    print("Reject Ho hypothesis")
else:
    print("Can not reject Ho hypothesis")
```

```python
from scipy import stats as st
import numpy as np
       
sample_1_2 = full_data.query('city == "Москва"').loc[:,'revenue'] 
sample_2_2 = full_data.query('city != "Москва"').loc[:,'revenue'] 

# check the variance of sample_1 and sample_2 to set equal_var = True
variance_1_2 = np.var(sample_1_2, ddof = 1)
variance_2_2 = np.var(sample_2_2, ddof = 1)
print('var_1: {:.3f}'.format(variance_1_2))
print('var_2: {:.3f}'.format(variance_2_2))
print('Difference of var1 and var_2 = {:.0%}'.format(variance_1_2 / variance_2_2))
print()

alpha = 0.05

results = st.ttest_ind(sample_1, sample_2, equal_var = False)
print('p-value: {:.3f}'.format(results.pvalue))

# Нулевая гепотеза. Ho: Средняя выручка По Москве равна средне выручке по остальным регионам
# Альтернативная гипотеза. H1: Средняя выручка По Москве не равна средней выручке по остальным регионам
# Zero hypothesis. Ho: Average revenue in Moscow is equal to average revenue in other regions
# Alternative hypothesis. H1: Average revenue in Moscow is not equal to average revenue in other regions

if (results.pvalue < alpha):
    print("Reject Ho hypothesis")
else:
    print("Can not reject Ho hypothesis")
```
