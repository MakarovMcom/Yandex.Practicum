 - Uid — уникальный идентификатор пользователя

Таблица `costs` (информация о затратах на маркетинг):
 - source_id — идентификатор рекламного источника
 - dt — дата
 - costs — затраты на этот рекламный источник в этот день

**Как часто люди возвращаются?**

```python
# make month cohorts
first_activity_date = visits_copy.groupby(['Uid'])['Start Ts'].min().reset_index()
first_activity_date.columns = ['Uid','first_activity_date']
visits_copy = visits_copy.merge(first_activity_date,on='Uid')
display(visits_copy.head())
```

```python
visits_copy['cohort_lifetime'] = (visits_copy['Start Ts'] - visits_copy['first_activity_date']) \
                            / np.timedelta64(1,'M')
visits_copy['cohort_lifetime'] = visits_copy['cohort_lifetime'].astype('int')
visits_copy['first_activity_month'] = visits_copy['first_activity_date'].astype('datetime64[M]')

cohorts = visits_copy.groupby(['first_activity_month','cohort_lifetime']) \
            .agg({'Uid':'nunique'}).reset_index()
cohorts.columns = ['first_activity_month','cohort_lifetime','uniq_users']
display(cohorts.head())
```

```python
#retention rate
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_activity_month','uniq_users']]
initial_users_count = initial_users_count.rename(columns={'uniq_users':'cohort_users'}) 
cohorts = cohorts.merge(initial_users_count,on='first_activity_month')
cohorts['retention'] = cohorts['uniq_users']/cohorts['cohort_users']
mean_retention = cohorts.query('cohort_lifetime != 0')['retention'].mean()
retention_pivot = cohorts.pivot_table(
    index = 'first_activity_month',
    columns = 'cohort_lifetime',
    values = 'retention',
    aggfunc = 'sum')
print('Медианный Retention Rate - {:.2%}'.format(mean_retention))
```

```python
# Median retention rate for cohorts without lifetime = 0

cohort_retention = retention_pivot.median(axis=0)
cohort_retention = cohort_retention.drop(0)
#cohort_retention = cohort_retention.map(lambda x: '{:.2%}'.format(x))
print(cohort_retention.map(lambda x: '{:.2%}'.format(x)))
cohort_retention.plot(kind = 'bar', grid = True, figsize = (12,6));
plt.title("Median retention rate for cohorts without lifetime = 0", fontsize = 20);
plt.xlabel("cohort lifetime [month]", fontsize = 15); # label for x-axis
plt.xticks(rotation=0);
plt.ylabel("Retention Rate", fontsize = 15); # label for y-axis

for i, val in enumerate(cohort_retention.values):
    plt.text(
             i, 
             val, 
             np.round(float(val),3), 
             horizontalalignment='center', 
             verticalalignment='bottom', 
             fontdict={'fontweight':500, 'size':12}
            )
```

```python
sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('Cohorts: User Retention')
sns.heatmap(retention_pivot, 
            annot=True, 
            fmt='.1%', 
            linewidths=1, 
            linecolor='gray',
            vmin = 0,
            vmax = 0.1,
            cmap = 'coolwarm',
            yticklabels=retention_pivot.index.strftime('%Y-%m')); 
```

**Выводы**:

   Средний Retention Rate = 3.76%. Слишком мало. Нужно подумать маркетологам, как удерживать пользователей. Тенденция такова, что пользователи уходят от сервиса.

**Сколько денег приносят? Вариант №1 (LTV)**

Когорта - источник трафика

```python
# определяем источник пользователя
sources = visits.sort_values(by = ['Uid', 'Start Ts']).groupby('Uid').agg({'Source Id': 'first'})
display(sources.head())
```

```python
# соединяем первые покупки и источники 
payers = orders.groupby('Uid').agg({'Buy Ts': 'min'}).merge(sources, on = 'Uid').reset_index()
payers = payers.rename(columns = {'Buy Ts': 'first_buy'})
display(payers.head())
```

```python
# считаем ltv
ltv = payers.merge(orders, on = 'Uid')
ltv['lifetime'] = ((ltv['Buy Ts'] - ltv['first_buy']) / np.timedelta64(1,'M')).round().astype(int)
display(ltv)
```

```python
# Тут делаем сводную таблицу чтобы посчитать валовую прибыль
report_ltv = ltv.groupby(['lifetime','Source Id']).agg({'Uid':'nunique', 'Revenue': 'sum' }).reset_index()
report_ltv.columns = ['lifetime', 'Source Id', 'n_buyers', 'Revenue']
report_ltv
```

```python
margin_rate = 1
report_ltv['gp'] = report_ltv ['Revenue'] * margin_rate
display(report_ltv.head())
```

```python
# Считаем и присоеденяем столбец с LTV
report_ltv['ltv'] = report_ltv['gp'] / report_ltv['n_buyers']
```

```python
output = report_ltv.pivot_table(
    index = 'Source Id', 
    columns = 'lifetime', 
    values='ltv', 
    aggfunc='mean').round().fillna('')
display(output)
```

```python
output_ltv_by_source = report_ltv.pivot_table(
    index = 'Source Id', 
    #columns = 'lifetime', 
    values='ltv', 
    aggfunc='mean').round().fillna(0)
display(output_ltv_by_source.cumsum(axis=1))
```

**Выводы:**

Когорта - первый месяц входа

```python
#LTV удобно считать по когортам. 
#найдем дату первой покупки каждого покупателя

orders_copy['orders_month'] = orders_copy['Buy Ts'].astype('datetime64[M]')
first_orders = orders_copy.groupby('Uid').agg({'orders_month': 'min'}).reset_index()
first_orders.columns = ['Uid', 'first_order_month']
display(first_orders.head()) 
```

```python
#вычислим число новых покупателей за каждую дату

cohort_sizes = first_orders.groupby('first_order_month').agg({'Uid': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_order_month', 'n_buyers']
display(cohort_sizes.head()) 
```

```python
#добавим дату первой покупки в таблицу с заказами

orders_new = pd.merge(orders_copy,first_orders, on='Uid')
display(orders_new.head())
```

```python
#сгруппируем таблицу и посчитаем выручку

cohorts_orders = orders_new.groupby(['first_order_month','orders_month']).agg({'Revenue': 'sum'}).reset_index()
display(cohorts_orders.head())
```

```python
#добавим в таблицу количество пользователей
report_revenue = pd.merge(cohort_sizes, cohorts_orders, on ='first_order_month')
display(report_revenue.head())
```

```python
#LTV

report_revenue['gp'] = report_revenue['Revenue'] *  margin_rate
report_revenue['age'] = (report_revenue['orders_month'] - report_revenue['first_order_month']) / np.timedelta64(1, 'M')
report_revenue['age'] = report_revenue['age'].round().astype('int')
report_revenue['ltv'] = report_revenue['gp'] / report_revenue['n_buyers']

#final step LTV in pivot table
ltv_final_pivot = np.round(report_revenue.pivot_table(
        index='first_order_month', 
        columns='age', 
        values='ltv', 
        aggfunc='mean'),2) 
display(ltv_final_pivot.fillna(''))
```

```python
cum_ltv = ltv_final_pivot.cumsum(axis=1)
```

```python
sns.set(style='white')
plt.figure(figsize=(19, 15))
plt.title('LTV')
sns.heatmap(cum_ltv, 
            annot=True, 
            #fmt='.1%', 
            linewidths=1, 
            linecolor='gray',
            vmin = 3,
            vmax = 12,
            cmap = 'coolwarm',
            yticklabels=ltv_final_pivot.index.strftime('%Y-%m'),
            #xticklabels=avg_price.index.strftime('%Y-%m')
           ); 
```

**Выводы:**
    
    В целом, месяц к месяцу пользователи приносят все меньше выручки.

```python
# CAC - сгруппировать costs по источникам и поделить на все то же количество покупателей 
# из каждого источника

# посчитаем количество покупателей из каждого источника
report_ltv = report_ltv.groupby('Source Id')['n_buyers'].sum().reset_index()
report_ltv = report_ltv.rename(columns = {"Source Id":'source_id'})
display(report_ltv)
```

**ROMI по рекламным источникам и времени**

ROI = LTV / CAC

LTV = (Revenue * margin_rate) / n_buyers

CAC = Costs / n_buyers

=> **ROI = Revenue * margin_rate / Costs**

margin_rate = 1 

=> **ROI = Revenue / Costs**

```python
# определяем источник пользователя
sources = visits.sort_values(by = ['Uid', 'Start Ts']).groupby('Uid').agg({'Source Id': 'first'})

# соединяем первые покупки и источники 
payers = orders.groupby('Uid').agg({'Buy Ts': 'min'}).merge(sources, on = 'Uid').reset_index()
payers = payers.rename(columns = {'Buy Ts': 'first_buy'})
payers
```

```python
# считаем ltv
ltv = payers.merge(orders, on = 'Uid')
ltv['lifetime'] = ((ltv['Buy Ts'] - ltv['first_buy']) / np.timedelta64(1,'M')).round().astype(int)
#orders_lifetime = ltv.pivot_table(index = 'Source Id', columns = 'lifetime', values = 'Revenue', aggfunc = 'count').fillna(0)
#buyers_lifetime = ltv.pivot_table(index = 'Source Id', columns = 'lifetime', values = 'Uid', aggfunc = 'nunique').fillna(0)
ltv = ltv.pivot_table(index = 'Source Id', columns = 'lifetime', values = 'Revenue', aggfunc = 'sum').fillna(0)
ltv = ltv.cumsum(axis = 1)
display(ltv)
```

```python
# добавляем размеры когорт и делим на них
# div(ltv['cohort_size'], axis = 0) берет столбец cohort_size и делит на него каждый столбец фрейма
cohort_sizes = payers.groupby('Source Id').agg({'Uid': 'nunique'}).rename(columns = {'Uid': 'cohort_size'})
display(cohort_sizes)
```

```python
ltv = ltv.merge(cohort_sizes, on = 'Source Id')
display(ltv)
```

```python
ltv = ltv.div(ltv['cohort_size'], axis = 0).drop(columns = 'cohort_size')
display(ltv)
```

```python
# считаем САС
costs = costs.rename(columns = {'source_id':'Source Id'})
cac = costs.groupby('Source Id').agg({'costs': 'sum'}).merge(cohort_sizes, on = 'Source Id')
#display(cac)
cac['cac'] = cac['costs'] / cac['cohort_size']
display(cac)
```

```python
# считаем ROI
roi = ltv.merge(cac[['cac']], on = 'Source Id')
roi = roi.div(roi['cac'], axis = 0).drop(columns = 'cac')
roi
```

```python
roi.T.plot(figsize = (17,10), linewidth = 2)
plt.axhline(y = 1, label = 'Уровень окупаемости', color = 'red', linewidth = 3, linestyle = '--')
plt.title('ROI по источникам привлечения')
plt.xlabel('Lifetime [month]')
plt.legend()
plt.show();
```

**Выводы:**
    
    На второй месяц быстро окупается первый Источник трафика, у остальных каналов с течением времени окупаемость слабая.
